```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import AdaBoostClassifier,GradientBoostingClassifier,StackingClassifier
from sklearn.metrics import accuracy_score,confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt 
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
```

```python
df=pd.read_csv("winequality-white.csv",sep=';')
df
```

```python
X=df.drop('quality',axis=1)
Y=df['quality']
```

```python
x_train,x_test,y_train,y_test=train_test_split(X,Y,train_size=0.8,random_state=42)
```

```python
svc=SVC(random_state=42)
lr=LogisticRegression(random_state=42)
knn=KNeighborsClassifier()
```

```python
from sklearn.tree import DecisionTreeClassifier
dt=DecisionTreeClassifier(random_state=42)
ADA=AdaBoostClassifier(base_estimator=dt,n_estimators=50,learning_rate=0.1,random_state=42)
GB=GradientBoostingClassifier(n_estimators=100,random_state=42,learning_rate=0.1)
STACKING=StackingClassifier(estimators=[('svc',svc),('dt',dt),('knn',knn)],final_estimator=lr)
```

```python
for clf in [ADA,GB,STACKING]:
    clf.fit(x_train,y_train)
    y_pred=clf.predict(x_test)
    print(f"{clf.__class__.__name__}:Accuracy is :{accuracy_score(y_test,y_pred)}")
```

```python
def plot_confusion_matrix(y_true,y_pred,title):
    con=confusion_matrix(y_true,y_pred,labels=range(3,9))
    sns.heatmap(con,cmap=plt.cm.Blues,fmt='d',annot=True,cbar=False)
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    plt.title(title)
    plt.show()
```

```python
for clf in [ADA,GB,STACKING]:
    clf.fit(x_train,y_train)
    y_pred=clf.predict(x_test)
    title=clf.__class__.__name__
    plot_confusion_matrix(y_test,y_pred,title)
```

```python
for clf in [ADA,GB,STACKING]:
    clf.fit(x_train,y_train)
    y_pred=clf.predict(x_test)
    plt.hist(y_pred,bins=range(3,9),alpha=0.5,label=clf.__class__.__name__)
    plt.xlabel('Predicted Quality')
    plt.ylabel('Frequency')
    plt.legend()
plt.show()
```
